### Two level of Pivoting the data

1st is pd.melt.  
melt is converting it to long format

```python
pfizer_melt = pd.melt(    pfizer, id_vars= ['Date', 'Drug_Name', 'Parameter'], var_name= "Time", value_name= "Reading" )
                                                                       
```

```python
pfizer_melt
```

```python
pfizer_melt.shape
```

2nd is pd.pivot

```python
df_final= pd.pivot(pfizer_melt, 
         index= ["Date", "Drug_Name", "Time"], 
         columns= "Parameter", 
         values= "Reading").reset_index()
```

```python
df_final
```

```python
df_final.shape
```

```python
df_final.columns.name = None
```

```python
df_final
```

We want to find the mean temperature and pressure  of drugs across dates 

```python
#default is mean
pd.pivot_table( df_final, 
               index= "Drug_Name", 
               columns= "Date",
               values= ["Temperature", "Pressure"])

```

We want to know the max temperature reached

```python
pd.pivot_table( df_final, 
               index= "Drug_Name", 
               columns= "Date",
               values= ["Temperature", "Pressure"],
               aggfunc= "max", 
               margins= True)
```

We want to know the min temperature reached

```python
pd.pivot_table( df_final, 
               index= "Drug_Name", 
               columns= "Date",
               values= ["Temperature", "Pressure"],
               aggfunc= "min", 
               margins= True)
```

lets replace the null values with mean temperature/pressure of the specific drug

```python
def cal_mean(x):
    x["Press_mean"] = x["Pressure"].mean()
    x["Temp_mean"] = x["Temperature"].mean()
    return x
```

```python
df_final.groupby(["Date", "Drug_Name"]).apply(cal_mean)


```
